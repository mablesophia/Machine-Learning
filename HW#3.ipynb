```python
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml

iris = np.genfromtxt("data/iris.txt", delimiter=None)
X, Y = iris[:, 0:2], iris[:, -1]
X, Y = ml.shuffleData(X, Y)
X, _ = ml.transforms.rescale(X)

XA, YA = X[Y<2, :], Y[Y<2]
XB, YB = X[Y>0, :], Y[Y>0]
```

```python
# problem 1(a)
ml.plotClassify2D(None, XA, YA)
plt.title("Graph for Data Set A")
plt.show()
```

```python
ml.plotClassify2D(None, XB, YB)
plt.title("Graph for Data Set B")
plt.show()
```

```python
# problem 1(b)
from logisticClassify2 import *

learner_A = logisticClassify2();
learner_A.classes = np.unique(YA)
wts = np.array([0.5, 1, -0.25])
learner_A.theta = wts

logisticClassify2.plotBoundary(learner_A, XA, YA)
plt.title("Actual Graph for Data Set A")
plt.show()
```

```python
learner_B = logisticClassify2();
learner_B.classes = np.unique(YA)
wts = np.array([0.5, 1, -0.25])
learner_B.theta = wts

logisticClassify2.plotBoundary(learner_B, XB, YB)
plt.title("Actual Graph for Data Set B")
plt.show()
```

```python
Yhat = logisticClassify2.predict(learner_A, XA);
ml.plotClassify2D(learner_A, XA, YA)
plt.title("Predict Graph for Data Set A")
plt.show()
```

```python
print("error rate for Data Set A",learner_A.err(XA, YA))
print("error rate for Data Set B",learner_B.err(XB, YB))
```
